```python
# questão 3
import random
pontos = [{"id": 1, "x":-1.198, "y":-5.164},{"id": 2, "x":5.573, "y":7.114},{"id": 3, "x":-6.614, "y":0.072},{"id": 4, "x":-7.374, "y":-1.107},{"id": 5, "x":-9.251, "y":8.321},{"id": 6, "x":6.498, "y":-6.036},{"id": 7, "x":0.861, "y":6.903},{"id": 8, "x":3.904, "y":-5.261},{"id": 9, "x":7.976, "y":-9.000}, {"id": 10, "x": -2.61, "y": 0.039},{"id": 11, "x": 4.487, "y":7.142},{"id": 12, "x": 8.938, "y": -4.388},{"id": 13, "x": -4.17, "y": -9.09},{"id": 14, "x": 7.835, "y": -9.269},{"id": 15, "x": 2.792, "y": -7.944},{"id": 16, "x": 5.212, "y":9.271},{"id": 17, "x": 6.687, "y": 6.731},{"id": 18, "x": -2.19, "y": -9.21},{"id": 19, "x": -1.06, "y":8.752},{"id": 20, "x": 6.883, "y":0.882}]
sub_lista = []
id = []
x = []
y = []
def gerador_pontos():
  intervalo1 = random.choice(pontos)
  intervalo2 = random.choice(pontos)
  if intervalo1['id'] < intervalo2['id']:
    print(f"Gerando dois pontos aletórios da Lista: {intervalo1, intervalo2}")
    indice_1 = intervalo1['id'] - 1
    indice_2 = intervalo2['id']
    for i in range(indice_1, indice_2, 1):
      sub_lista.append(pontos[i])
    print(f"\nSub-lista = {sub_lista}")
    random.shuffle(sub_lista)
    print(f"\nSub-lista embalharada = {sub_lista}\n")
    for i in sub_lista:
      plt.scatter(i['x'],i['y'], marker='^', s=50)
    plt.title('Coordenadas dos pontos')
    plt.xlabel('Eixo X')
    plt.ylabel('Eixo y')
    plt.legend(sub_lista, bbox_to_anchor = (1.05, 1.03), title = "Legenda:")
    plt.show()
    for i in sub_lista:
      id.append(i['id'])
      x.append(i['x'])
      y.append(i['y'])
    plt.plot(x, y, '^', color='black')
    plt.plot(x, y, 'k--', color='blue')
    plt.title('Pontos conectados')
    plt.xlabel('Eixo X')
    plt.ylabel('Eixo y')
    plt.show()
  elif intervalo1['id'] == intervalo2['id'] or intervalo1['id'] > intervalo2['id']:
    gerador_pontos() 
gerador_pontos()
```
